```python
import random

population = list(range(1, 301))
sample_size = 10
random_sample = random.sample(population , sample_size)
print("Random Sample : ", random_sample)
```

```python
import numpy as np
import pandas as pd

data = pd.DataFrame({
    'Category' : ['A' , 'B' , 'C' , 'D'] * 25,
    'Value' : np.random.randn(100)
})

stratified_Sample = data.groupby('Category' , group_keys=False).apply(lambda x: x.sample(2))
print("Stratified Sample:\n" , stratified_Sample)
```

```python
#Cluster Sampling

clusters = [list(range(1, 26)), list(range(26, 51)), list(range(51, 76)), list(range(21, 88))]
cluster_sample = random.choice(clusters)
print("Clusters Sample:", cluster_sample)
```

```python
#Data Summarization

import pandas as pd
import numpy as np

data = pd.DataFrame({
    'Values' : np.random.randn(100)
})

summary = data.describe()
print("Summary Statistics:\n", summary)
```

```python
#Data Visualisation

import matplotlib.pyplot as plt

data['Values'].hist(bins=10, edgecolor='black')
plt.title('Histogram of Values')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# Load the iris dataset
iris = load_iris()

# Apply PCA to reduce the data to 2 dimensions
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(iris.data)

# Scatter plot of the reduced data
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=iris.target, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Iris Dataset')
plt.colorbar(label='Species')
plt.show()
```
